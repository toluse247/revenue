```python
import pandas as pd
import numpy as np
```

## Import csv data files

```python
billing = pd.read_csv("./data/billing_details.csv", dtype=object)
```

```python
billing.head()
```

```python
teams = pd.read_csv("./data/sales_teams.csv", dtype=object)
```

```python
teams.head()
```

```python
payments = pd.read_csv("./data/payment_records.csv", dtype=object)
```

```python
payments.head()
```

Find out dtypes of the columns of "billing", "payments" and "teams"

```python
billing.dtypes
```

```python
payments.dtypes
```

```python
teams.dtypes
```

Select only relevant columns from billing

```python
billing = billing[["businessunitname", "serviceunitname", "globalaccountnumber", "name", "serviceaddress", "customertype", "displaycode", "billedamount", "outstandingamount", "billyear", "updateddssname"]]
```

```python
billing
```

#### Rename columns to conform with "payment records" naming convention and for clearer understanding

```python
billing.rename(columns={"businessunitname" : "district", "globalaccountnumber" : "account_no", "serviceaddress" : "address", "displaycode":"bill_type", "updateddssname":"transformer"}, inplace=True)
```

```python
billing.head()
```

Convert account number, billed amount, outstanding amount to numerical values


```python
billing.account_no = billing.account_no.apply(pd.to_numeric, errors = "coerce").fillna(0)
```

```python
billing.outstandingamount = billing.outstandingamount.apply(pd.to_numeric, errors = "coerce").fillna(0)
```

```python
billing.billedamount = billing.billedamount.apply(pd.to_numeric, errors = "coerce").fillna(0)
```

```python
billing.dtypes
```

Convert amount in 'payments' to numerical values

```python
payments.amount = payments.amount.apply(pd.to_numeric, errors = "coerce").fillna(0)
```

```python
payments.dtypes
```

Group by account_no to capture accounts with multiple payments within the same period

```python
payment_sum = payments.groupby("account_no")["amount"].sum()
```

```python
payment_sum
```

Convert the Series into a DataFrame

```python
payment_sum = payment_sum.reset_index()
```

```python
payment_sum
```

Rename "amount" column to "payment"

```python
 payment_sum.rename(columns={"amount": "payment"}, inplace=True)
```

```python
payment_sum
```

Convert account number to numerical values

```python
payment_sum.account_no = payment_sum.account_no.apply(pd.to_numeric, errors="coerce").fillna(0)
```

```python
payment_sum.dtypes
```

Merge Billing and Payment records, fetch "payment" column using customer account number

```python
performance = pd.merge(billing, payment_sum, on="account_no", how = "left").fillna(0)
```

```python
performance.head()
```

Select relevant columns from 'teams'

```python
teams = teams[["transformer", "marketer", "team_leader"]]
```

```python
teams.head()
```

Merge Performance data with sales teams

```python
performance = pd.merge(performance, teams, on="transformer", how = "left").fillna(0)
```

```python
performance.head()
```

Export to CSV

```python
performance.to_csv("performance.csv", index=False)
```
